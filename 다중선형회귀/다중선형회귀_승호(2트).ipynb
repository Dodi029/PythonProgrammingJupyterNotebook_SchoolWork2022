```python
f = open('./data.txt','r',encoding='utf-8')
data = f.readlines()
    
f.close()
```

```python
import numpy as np

class multipul_leaner_regression():
    def __init__(self, w = 0, inter = 1):
        self.w = w
        self.inter = inter
    
    def file(self, data):
        
        xy_list = []

        for i in range (len(data)):
            xy_list.append(data[i].split('\t'))

        for i in range(1, len(xy_list)):
        
            xy_list[i][1] = int(xy_list[i][1])
            xy_list[i][2] = int(xy_list[i][2])
            xy_list[i][3] = int(xy_list[i][3])
            xy_list[i][4] = int(xy_list[i][4])
            xy_list[i][5] = int(xy_list[i][5])
            xy_list[i][6] = int(xy_list[i][6])
            if self.inter == 1:
                xy_list[i].append(1)
    
        data_x1 = []
        data_x2 = []
        data_x3 = []
        data_x4 = []
        data_x5 = []
        data_x = []
        data_y = []

        for i in range(1,len(xy_list)):
            data_y.append(xy_list[i][1])
            data_x1.append(xy_list[i][2])
            data_x2.append(xy_list[i][3])
            data_x3.append(xy_list[i][4])
            data_x4.append(xy_list[i][5])
            data_x5.append(xy_list[i][6])
            
            if self.inter == 1:
                data_x.append([xy_list[i][1], xy_list[i][2], xy_list[i][3], xy_list[i][4], xy_list[i][5], xy_list[i][6], xy_list[i][7]])   
            else:
                data_x.append([xy_list[i][1], xy_list[i][2], xy_list[i][3], xy_list[i][4], xy_list[i][5], xy_list[i][6]])
    
        x_array = np.array(data_x)

        rand = np.random.RandomState(100)
        rand.shuffle(x_array)
        
        return x_array

    def slicing(self, x_array, p):
        
        tr = round(p * len(x_array))
        x_tr = []
        x_ts = []
        y_tr = []
        y_ts = []
    
        if self.inter == 1:
            n = 7
        else:
            n = 6
            
        x_tr = x_array[:tr, 1:n]
        x_tr = np.array(x_tr)

        x_ts = x_array[tr:, 1:n]
        x_ts = np.array(x_ts)
    
        for i in range(tr):
            y_tr.append([x_array[i][0]])
        y_tr = np.array(y_tr)

        for i in range(len(x_array)-tr):
            y_ts.append([x_array[i+tr][0]])
        y_ts = np.array(y_ts)
        
        return x_tr, x_ts, y_tr, y_ts 
    
    def ㅣearn(self, x ,y):
        self.w = np.linalg.inv(x.T @ x) @ x.T @ y
    
    def test(self, x, y):
        pred_y = []
        pred_y = np.array(pred_y)
        pred_y = np.dot(x, self.w)
        print("예측값 | 실제값")
        for i in range(len(pred_y)):
            print("%.1f"%(pred_y[i]), y[i])
        return pred_y
            
class Metric:
    def MSE(self, pred_y, y):
        mse_e = sum((y - pred_y)**2)/len(y)
        return mse_e
    
    def MAPE(self, pred_y, y):
        mape_e = sum(abs(y - pred_y)/abs(y))/len(y) * 100
        return mape_e
```

```python
l = multipul_leaner_regression()
m = Metric()

p = 0.75 #(75 : 25)

data_pre = l.file(data) #전처리
slicing = l.slicing(data_pre, p) #training test파일 분할
l.ㅣearn(slicing[0], slicing[2]) #w값 예측
pred_y = l.test(slicing[1], slicing[3]) #예측값과 실제값 비교
mse = m.MSE(pred_y,slicing[3]) #MSE
print("MSE :", mse)
mape = m.MAPE(pred_y,slicing[3]) #MAPE
print("MAPE :", mape)
```
