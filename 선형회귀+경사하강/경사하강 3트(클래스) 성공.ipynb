```python
import os
import math
import matplotlib.pyplot as plt
import numpy as np
os.getcwd()

f = open('/Users/doyoung/Downloads/Simple_Linear_Regression_Practice_Data.txt', 'r', encoding='utf-8')
data = f.readlines()
f.close()

xy = []

for i in range (len(data)):
    data[i]=data[i].replace(',', '.')
    #data[i]=data[i].replace('""', ' ')
    xy.append(data[i].split('\t'))
    
for i in range(1, len(xy)):
    xy[i][1]=xy[i][1].replace('"','')
    
for i in range(1, len(xy)):
    xy[0][1] = xy[0][1].rstrip%()
    xy[i][0] = int(xy[i][0])
    xy[i][1] = float(xy[i][1])

data_x=[]
data_y=[]

for i in range(1,len(xy)):
    data_x.append(xy[i][0])
    data_y.append(xy[i][1])
    
x_data = np.array(data_x) #넘파이 배열로 변환
y_data = np.array(data_y)

w = 1 # 기울기 초기화
w1= 1 # y절편 초기화

tr = 0.0005 # 학습률

ep = 4000 # 반복횟수
error = 0

class gradient_descent:
    def learn(self, x_data, y_data, error):
        y_pred = w * x_data + w1 # y 예측 값을 구하는 식
        error = y_data - y_pred # 오차 error = y 값 - y 예측 값
    
        w_diff = -(1 / len(x_data)) * sum(x_data * (error)) # 평균 제곱 오차를 w로 미분한 값
        w1_diff = -(1 / len(x_data)) * sum(y_data - y_pred) # 평균 제곱 오차를 w1으로 미분한 값
            
        return w_diff, w1_diff, error

    def pred(self, w_diff, w1_diff, tr):
        global w,w1
        w = w - tr * w_diff
        w1 = w1 - tr * w1_diff
        
        return w, w1
    
gd = gradient_descent()
for i in range(ep):
    gd_learn = gd.learn(x_data, y_data, error)
    gd_pred = gd.pred(gd_learn[0], gd_learn[1],tr)
    if i%100 == 0:    
        print("반복 = %.f, w = %.04f, w1 = %.04f, 오류 = %.04f" %(i, gd_pred[0], gd_pred[1], gd_learn[2].mean()))
```
