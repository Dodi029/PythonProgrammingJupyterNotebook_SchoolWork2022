```python
#xy_list는 x,y가 들어있는 리스트를 모아둔 이중리스트
x_data_list = []
y_data_list = []

for i in range(1, len(xy_list)):
    x_data_list.append(xy_list[i][0])
    y_data_list.append(xy_list[i][1])

import matplotlib.pyplot as plt
from random import random

plt.figure(figsize=(5,5))
plt.scatter(x_data_list,y_data_list)
plt.show()

def sigmaAll(list):
    total = 0
    for num in list:
        total += num
    return total

def trainingData(data_list, training_range):
    training_data_list = []
    for i in range(0, int(len(data_list) * training_range)):
        training_data_list.append(data_list[i])
    return training_data_list

def testData(data_list, training_data):
    for num in training_data:
        data_list.remove(num)
    return data_list

def deviationData(data_list, avg):
    dev_list = []
    for num in data_list:
        dev_list.append(num-avg)
    return dev_list

tr_range = 0.7
x_tr_data = trainingData(x_data_list, tr_range)
y_tr_data = trainingData(y_data_list, tr_range)
x_ts_data = testData(x_data_list, x_tr_data)
y_ts_data = testData(y_data_list, y_tr_data)

x_avg = sigmaAll(x_tr_data) / len(x_tr_data)
y_avg = sigmaAll(y_tr_data) / len(y_tr_data)

class Linear:
    
    def learn(self, x_tr_data, y_tr_data):
        x_avg = sigmaAll(x_tr_data) / len(x_tr_data)
        y_avg = sigmaAll(y_tr_data) / len(y_tr_data)
        
        x_dev_list = deviationData(x_tr_data, x_avg)
        y_dev_list = deviationData(y_tr_data, y_avg)
        
        dev_mul_list = []
        x_pow_list = []
        
        for i in range(0, len(x_dev_list)):
            dev_mul_list.append(x_dev_list[i] * y_dev_list[i])
            x_pow_list.append(x_dev_list[i]**2)
            
        w1 = sigmaAll(dev_mul_list) / sigmaAll(x_pow_list)
        w0 = y_avg - w1 * x_avg
        
        return w1,w0
        
    def predict(self, x_ts_data, y_ts_data, w0, w1):
        predict_data_list = []
        err_list = []
        err_squared_list = []
        
        for i in range(0, len(x_ts_data)):
            predict_data_list.append(w1 * x_ts_data[i] + w0)
            err_list.append(y_ts_data[i] - (w1*x_ts_data[i]+w0))
            err_squared_list.append((y_ts_data[i] - (w1*x_ts_data[i]+w0))**2)
        
        sum_err = sigmaAll(err_list)
        sum_err_squared = sigmaAll(err_squared_list)
        
        return sum_err/len(y_ts_data), sum_err_squared/len(y_ts_data)
    
    def status(self, w1, w0, sum_err, sum_err_squared):
        print("[선형회귀 결과]")
        print('w1 : {}'.format(w1))
        print('w0 : {}'.format(w0))
        print('err : {}'.format(sum_err))
        print('fin : {}'.format(sum_err_squared))
```
