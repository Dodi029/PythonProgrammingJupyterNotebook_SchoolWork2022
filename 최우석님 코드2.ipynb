```python
class MLR:
    def __init__(self,intercept=True):
        self.coef = None
        self.intercept = intercept
    def intercept_add(self,X):
        ones = np.ones(len(X)).reshape(-1,1)
        result = np.concatenate([ones,X],axis=1)
        return result
    def fit(self,X,Y):
        if self.intercept_add ==True:
            new_X = self.intercept_add(X)
            self.coef = np.linarg.inv(new_X.transpose()@new_X)@new_X.transpose()@Y
        else:
            self.coef = np.linarg.inv(X.transpose()@new_X)@X.transpose()@Y
    def predict(self,X):
        if self.intercept_add==True:
            new_X = self.intercept_add(X)
            result = new_X@self.coef
            return np.array(result)
        else:
            result = X@self.coef
            return np.array(result)
```
