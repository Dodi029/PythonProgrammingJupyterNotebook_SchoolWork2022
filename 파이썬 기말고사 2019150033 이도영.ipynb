```python
def split(X, Y, test_size=0.25):
    X = np.array(X)
    Y = np.array(Y)
    
    indices = np.array(range(len(X)))
    test_size_len = round(test_size * len(X))
    
    train_indices = indices[0:(len(X)-test_size_len)]
    test_indices = indices[(len(X)-test_size_len):]
        
    X_train = X[train_indices, :]
    X_test = X[test_indices, :]
    Y_train = Y[train_indices]
    Y_test = Y[test_indices]
        
    return X_train, X_test, Y_train, Y_test
```

```python
class MLR:
    def init(self,intercept=True):
        self.coef=None
        self.intercept=intercept
        
    def intercept_add(self,X_input):
        ones=np.ones(X_input.shape[0]).reshape(-1,1)
        result=np.concatenate([ones,X_input],axis=1)
        
        return result
        
    def fit(self,X_input, Y_output):
        if self.intercept==True:
            new_X_input=self.intercept_add(X_input)
            self.coef=np.linalg.inv(new_X_input.transpose()@new_X_input)@new_X_input.transpose()@Y_output
            
        else:
            new_X_input=X_input
            self.coef=np.linalg.inv(new_X_input.transposespose()@new_X_input)@new_X_input.transpose()@Y_output
            
    def predict(self,X_test):
        if self.intercept==True:
            new_X_test=self.intercept_add(X_test)
            pred_test=new_X_test@self.coef.reshape(-1,1)
            
        else:
            new_X_test=X_test
            pred_test=new_X_test@self.coef.reshape(-1,1)
            
        return pred_test
```

```python
class KNN:
    
    def init(self, k = 5):
        self.k = k
        self.neighbors = None
        
    def euclidean(self, obs1, obs2):
        dist = np.sqrt(np.sum(obs1-obs2)**2)
        
        return np.around(dist,3)
    
    def manhanttan(self, obs1, obs2):
        
        return np.sum(np.abs(obs1 - obs2))
        
    def search_neighbors(self, X_train, test_sample):
        dists_info=list()
        
        for index, train_sample in enumerate(X_train):
            dist = self.manhanttan(train_sample, test_sample)
            dists_info.append((train_sample,dist,index))
            
        dists_info.sort(key=lambda tupe:tupe[1])
        self.neighbors = dists_info[:self.k]
        
        return self.neighbors
    
    def predict(self, X_train,Y_train, test_sample):
        self.neighbors = self.search_neighbors(X_train, test_sample)
        
        neigh_index = list()
        
        for neigh in self.neighbors:
            neigh_index.append(neigh[2])
            
        neigh_y = Y_train[neigh_index]
        
        classes, counts = np.unique(neigh_y, return_counts = True)
        pred=classes[np.argmax(counts)]
        
        return pred,neigh_y
```

```python
def MSE(true_y, pred_y):
    result1=true_y - pred_y
    mse_result=np.mean(result1**2)
    
    return mse_result
    
def MAPE(true_y,pred_y):
    result2=true_y-pred_y
    result_mape=(np.abs((result2)/true_y))*100
    mape_result=np.mean(result_mape)
        
    return mape_result
```

```python
def Min_Max_scale(input):
    X_scaled=[]
    
    X_max=max(input)
    X_min=min(input)
    
    for i in range(0,len(input)):
        X_scaled.append((input[i]-X_min)/(X_max-X_min))
    
    return X_scaled
```
